# Remove ;

```python
df['teks'] = row
```

```python
line_raw =  df['teks']
rh =[]

for value in line:
  try:
    rh.append(re.sub(r'\n',' ',value))
  except:
    continue
```

```python
df['teks'] = rh
```

# Remove Emoticon

```python
def remove_emojis(text):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', text)
```

```python
df['data'] = df['teks'].apply(remove_emojis)
```

# Regular Expression

```python
def remove_link(text):
  # text = text.replace('\n',' ')
  # text = text.strip('\n')
  # text = re.sub(r'\n', ' ', text)
  text = re.sub(r'\d+', '', str(text))  
  text = re.sub(r'@\w+ ?|http\S+|#', '', str(text))
  text = re.sub(r"n't\s*|don$", "", str(text))
  text = re.sub(r"b'", '', str(text))
  text = re.sub(r"RT", '', str(text))
  text = re.sub(r':', '', str(text))
  text = re.sub(r'/', ' ', text)
  text = re.sub(r'[^\w\s]', ' ', str(text))
    
  # text = re.sub(r'\n ', '', str(text))
  # text = re.sub(r'x ', ' ', str(text))
  # text = re.sub(r'xf ', ' ', str(text))
  # text = re.sub(r'xe ', ' ', str(text))
  # text = re.sub(r'xc ', ' ', str(text))
  # text = re.sub(r'xa ', ' ', str(text))
  # text = re.sub(r'\n ', ' ', str(text))
  # text = re.sub(r'xb ', ' ', str(text))
  # text = re.sub(r'rp ', ' ', str(text))
  # text = re.sub(r'xad ', ' ', str(text))
  text = text.lower()

  return text
df2['data'] = df2['data'].apply(remove_link)
```

# Tokenizing

```python
regex_str = [
            #  r'\n',
    r'<[^>]+>',  # HTML tags
    r'(?:@[\w_]+)',  # @-mentions
    r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)",  # hash-tags
    r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+',  # URLs
    r'(?:(?:\d+,?)+(?:\.?\d+)?)',  # numbers
    r"(?:[a-z][a-z'\-_]+[a-z])",  # words with - and '
    r'(?:[\w_]+)',  # other words
    r'(?:\S)'  # anything else
]
tokens_re = re.compile(r'('+'|'.join(regex_str)+')', re.VERBOSE | re.IGNORECASE)

def tokenize(s):
    return tokens_re.findall(s)

df2['remove_regex'] = df2['data'].apply(tokenize)
```

# Join Text

```python
def joinn(text):
  text = " ".join(text)
  return text
df2['remove_regex'] = df2['remove_regex'].apply(joinn)
```

```python
import swifter
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
stopwords_indonesia = stopwords.words('indonesian')
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
from nltk import word_tokenize,sent_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nlp_id.lemmatizer import Lemmatizer
```

### With Sastrawi

```python
# get list of stopwords
stopword_list = StopWordRemoverFactory().get_stop_words()

# create stopwords instance
factory = StopWordRemoverFactory()
stopwords = factory.create_stop_word_remover()

# create a new list of stopwords
new_stopwords = ['x', 'y', 'r', 'e', 's', 'm', 'hi', 'yet', 'may', 'oh', 'due', 'to',
                'b', 'c', 'f', 'n', 'xde', 'co', 'xad',
                'way', 'wa', 'xe', 'xa', 'xb', 'xef', 'xf', 'xd', 'xba', 'xbb', 'xbc', 'xef', 'xc', 'xll','rt']
# add both stopwords 
all_stopwords = stopwords_indonesia + new_stopwords

# create ArrayDictionary 
stopword_dict = ArrayDictionary(all_stopwords)

sastrawi_stopwords = StopWordRemover(stopword_dict)

# create instance stopword

def remove_stopword_with_sastrawi(text):
    text = sastrawi_stopwords.remove(text)
    return text
```

```python
df2['stopword'] = df2['remove_regex'].apply(remove_stopword_with_sastrawi)
```

### With NLP

```python
def remove_stopword_with_nlp(text):
    # text =  stopword.remove_stopword(text)
    stop_words = set(stopwords_indonesia)
    word_tokens = word_tokenize(text)
    filtered_sentence = [w for w in word_tokens if not w in stop_words]
    return ' '.join(filtered_sentence)
    return text
```

# Steming

```python
def text_steming(text):
    # create stemmer
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    text = stemmer.stem(text)
    return text
```

# Lematizing

```python
lemmatizer = Lemmatizer()
def text_lemmatizing(text):
    text = lemmatizer.lemmatize(text)
    return text
```

# Main Preprocessing

```python
def preprocessing(text):
  text = remove_stopword_with_nlp(text)
  text = text_steming(text)
  text = text_lemmatizing(text)

  return text
```

```python
df2['hasil'] = df2['stopword'].swifter.apply(preprocessing)
```
